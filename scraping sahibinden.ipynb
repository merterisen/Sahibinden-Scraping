```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service

driverpath = ""
service = Service(driverpath)
driver = webdriver.Chrome(service=service)
website = "https://www.sahibinden.com/"
driver.get(website)


#arama
sahibinden_arama = "arsa"
driver.find_element(By.XPATH, '//*[@id="searchText"]')\
    .send_keys(sahibinden_arama)

driver.find_element(By.XPATH, '//*[@id="searchSuggestionForm"]/button')\
    .click()


# 50 sonuç göster
driver.execute_script("arguments[0].click();", WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="searchResultsSearchForm"]/div/div[3]/div[3]/div[2]/ul/li[2]/a'))))


#ürün bilgileri
item_titles = driver.find_elements(By.XPATH, '//*[@id="searchResultsTable"]/tbody/tr[*]/td[2]')
item_prices = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[3]')
item_loca = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[5]')
item_date = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[4]')


# boş liste
titles_list = []
prices_list = []
loca_list = []
date_list = []


# Loop over the item_titles and item_prices
for i in range(len(item_titles)):
    title = driver.find_elements(By.XPATH, '//*[@id="searchResultsTable"]/tbody/tr[*]/td[2]')[i]
    titles_list.append(title.text)
    
for i in range(len(item_prices)):
    price = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[3]')[i]
    prices_list.append(price.text)
    
for i in range(len(item_loca)):
    loca = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[5]')[i]
    loca_list.append(loca.text)
    
for i in range(len(item_date)):
    dat = driver.find_elements(By.XPATH,'//*[@id="searchResultsTable"]/tbody/tr[*]/td[4]')[i]
    date_list.append(dat.text)


driver.close()
```

```python
import pandas as pd

df = pd.DataFrame(zip(titles_list, prices_list, loca_list, date_list), columns=['İlan Başlığı', 'Fiyat', 'Lokasyon', 'Tarih'])
df['Fiyat'] = df['Fiyat'].str.replace(' TL', '')
df['Lokasyon'] = df['Lokasyon'].str.replace('\n', ' ')
df['Tarih'] = df['Tarih'].str.replace('\n', ' ')
df.drop('İlan Başlığı', axis=1, inplace=True)
df.head()
```
